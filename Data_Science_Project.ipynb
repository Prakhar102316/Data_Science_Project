```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_csv(r"C:\Users\cool7\Desktop\Football_player\players_15.csv")
print("Data loaded successfully!")
```

```python
# Displaying the attributes (columns) and data types
print("--- File Attributes ---")
print(df.columns.tolist())

print("\n--- Dataset Summary ---")
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")

# Showing basic info about null values and types
df.info()
```

```python
# Check for duplicates
initial_count = len(df)
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

# Drop them
df = df.drop_duplicates()
print(f"Rows remaining after dropping duplicates: {len(df)}")
```

```python
plt.figure(figsize=(12, 6))
top_nationalities = df['nationality'].value_counts().head(10)

sns.barplot(x=top_nationalities.index, y=top_nationalities.values, palette='magma')
plt.title('Top 10 Countries by Player Count')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
# Taking top 7 for a clean pie chart
pie_data = df['nationality'].value_counts().head(7)

plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Percentage Distribution of Top 7 Nationalities')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=20, kde=True, color='teal')
plt.title('Player Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
